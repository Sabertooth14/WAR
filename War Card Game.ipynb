```python
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
#CARD CLASS
class Card():
    
    def __init__(self,suits,ranks):
        
        self.suits = suits
        self.ranks = ranks
        self.values = values[ranks]
        
    def __str__(self):
        
        return self.ranks + " of " + self.suits
```

```python
#DECK CLASS
class Deck():
    
    def __init__(self):
        self.all_cards = []
        
        for s in suits:
            for r in ranks:
                my_card = Card(s,r)
                self.all_cards.append(my_card)
    
    def shuffled(self):
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        return self.all_cards.pop()
```

```python
#CLASS PLAYER

class Player:
    
    def __init__(self,name):
        self.name = name
        self.player_cards = []
        
    def remove_one(self):
        return self.player_cards.pop(0)
        
    def add_cards(self,new_cards):
        if type(new_cards)==type([]):
            #to add multiple cards
            self.player_cards.extend(new_cards)
            
        else:
            #to add a single card
            self.player_cards.append(new_cards)
    
    def __str__(self):
        return f"{self.name} has {len(self.player_cards)} cards"
```

```python
#CREATING PLAYERS
player_1 = Player("One")
player_2 = Player("Two")

#CREATIG DECK
new_deck = Deck()
new_deck.shuffled()

#DISTRIBUTING THE CARDS
for i in range(26):
    player_1.add_cards(new_deck.deal_one())
    player_2.add_cards(new_deck.deal_one())
```
